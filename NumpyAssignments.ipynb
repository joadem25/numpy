Using the BicycleWeather.csv dataset perform the following operations: 
    
    1. Extract the PRCP column from the pandas object. 
    2. Using boolean functions/operators separate numbers from NAN 
    3. Using boolean masking separate all zeros and non-zeros into different variables. 
    4. From the variable holding non-zeros, multiply every values in the array with 100 and divid by 250 
    5. Calculate the following: a. Median precip on rainy days b. Median precip on summer days c. Maximum precip on summer days d. Median precip on non-summer rainy days


```python
import numpy as np
import pandas as pd
```

```python
weather = pd.read_csv('C:\\PythonClass\\Data science\\Datasets\\BicycleWeather.csv')
weather_2012 = weather[0:366]
weather_2013 = weather[366:731]
weather_2014 = weather[731:1096]
weather_2015 = weather[1096:]
weather_2015.tail()
```

```python
# 1. Extract the PRCP column from the pandas object
prcp_2012 = np.array(weather_2012['PRCP'])
prcp_2013 = np.array(weather_2013['PRCP'])
prcp_2014 = np.array(weather_2014['PRCP'])
prcp_2015 = np.array(weather_2015['PRCP'])
print('PRCP (2012)\n', prcp_2012, '\n')
print('PRCP (2013)\n', prcp_2013, '\n')
print('PRCP (2014)\n', prcp_2014, '\n')
print('PRCP (2015)\n', prcp_2015)
# len(prcp_2015)
```

```python
# 2. Using boolean functions/operators separate numbers from NAN
prcp_2012_numbers = np.greater_equal(prcp_2012, 0)
prcp_2012_numbers = prcp_2012[prcp_2012_numbers]

prcp_2013_numbers = np.greater_equal(prcp_2013, 0)
prcp_2013_numbers = prcp_2013[prcp_2013_numbers]

prcp_2014_numbers = np.greater_equal(prcp_2014, 0)
prcp_2014_numbers = prcp_2014[prcp_2014_numbers]

prcp_2015_numbers = np.greater_equal(prcp_2015, 0)
prcp_2015_numbers = prcp_2015[prcp_2015_numbers]
print('PRCP numbers (2012)\n', prcp_2012_numbers, '\n')
print('PRCP numbers (2013)\n', prcp_2013_numbers, '\n')
print('PRCP numbers (2014)\n', prcp_2014_numbers, '\n')
print('PRCP numbers (2015)\n', prcp_2015_numbers)
```

```python
# 3. Using boolean masking separate all zeros and non-zeros into different variables.
# YEAR 2012
prcp_2012_nonzeros = np.greater(prcp_2012_numbers, 0)
prcp_2012_nonzeros = prcp_2012_numbers[prcp_2012_nonzeros]

prcp_2012_zeros = np.equal(prcp_2012_numbers, 0)
prcp_2012_zeros = prcp_2012_numbers[prcp_2012_zeros]

print('Non-Zeros YEAR 2012\n', prcp_2012_nonzeros, '\n')
print('Zeros YEAR 2012\n', prcp_2012_zeros, '\n\n')

# YEAR 2013
prcp_2013_nonzeros = np.greater(prcp_2013_numbers, 0)
prcp_2013_nonzeros = prcp_2013_numbers[prcp_2013_nonzeros]

prcp_2013_zeros = np.equal(prcp_2013_numbers, 0)
prcp_2013_zeros = prcp_2013_numbers[prcp_2013_zeros]

print('Non-Zeros YEAR 2013\n', prcp_2013_nonzeros, '\n')
print('Zeros YEAR 2013\n', prcp_2013_zeros, '\n\n')

# YEAR 2014
prcp_2014_nonzeros = np.greater(prcp_2014_numbers, 0)
prcp_2014_nonzeros = prcp_2014_numbers[prcp_2014_nonzeros]

prcp_2014_zeros = np.equal(prcp_2014_numbers, 0)
prcp_2014_zeros = prcp_2014_numbers[prcp_2014_zeros]

print('Non-Zeros YEAR 2014\n', prcp_2014_nonzeros, '\n')
print('Zeros YEAR 2014\n', prcp_2014_zeros, '\n\n')

# YEAR 2015
prcp_2015_nonzeros = np.greater(prcp_2015_numbers, 0)
prcp_2015_nonzeros = prcp_2015_numbers[prcp_2015_nonzeros]

prcp_2015_zeros = np.equal(prcp_2015_numbers, 0)
prcp_2015_zeros = prcp_2015_numbers[prcp_2015_zeros]

print('Non-Zeros YEAR 2015\n', prcp_2015_nonzeros, '\n')
print('Zeros YEAR 2015\n', prcp_2015_zeros)
```

```python
# 4. From the variable holding non-zeros, multiply every values in the array with 100 and divid by 250
prcp_inches_2012 = prcp_2012_nonzeros*(100/250)
prcp_inches_2013 = prcp_2013_nonzeros*(100/250)
prcp_inches_2014 = prcp_2014_nonzeros*(100/250)
prcp_inches_2015 = prcp_2015_nonzeros*(100/250)
print('PRCP value in inches (2012)\n', prcp_inches_2012, '\n')
print('PRCP value in inches (2013)\n', prcp_inches_2013, '\n')
print('PRCP value in inches (2014)\n', prcp_inches_2014, '\n')
print('PRCP value in inches (2015)\n', prcp_inches_2015)
```

```python
#  5. Calculate the following: a. Median precip on rainy days b. Median precip on summer days 
#     c. Maximum precip on summer days d. Median precip on non-summer rainy days

# a. Median precip on rainy days
median_2012 = np.median(prcp_inches_2012)
median_2013 = np.median(prcp_inches_2013)
median_2014 = np.median(prcp_inches_2014)
# median_2015 = np.median(prcp_inches_2015) # Year 2015 is not a complete data
print('The Median on rainy days is (2012): ', median_2012)
print('The Median on rainy days is (2013): ', median_2013)
print('The Median on rainy days is (2014): ', median_2014)

# b. Median precip on summer days 
# Summer starts from the 172nd day and last for about 90 days
summer_leap_year = (np.arange(366)-171<90) & (np.arange(366)-171>=0)
summer = (np.arange(365)-171<90) & (np.arange(365)-171>=0)

median_2012_summer = np.median(prcp_2012_numbers[summer_leap_year])
median_2013_summer = np.median(prcp_2013_numbers[summer])
median_2014_summer = np.median(prcp_2014_numbers[summer])
# median_2015_summer = np.median(prcp_2015_numbers[summer]) # Year 2015 is not a complete data 
print('The Median on summer days is (2012): ', median_2012_summer*(100/250))
print('The Median on summer days is (2013): ', median_2013_summer*(100/250))
print('The Median on summer days is (2014): ', median_2014_summer*(100/250))

# c. Maximum precip on summer days
max_prcp_2012_summer = np.max(prcp_2012_numbers[summer_leap_year])
max_prcp_2013_summer = np.max(prcp_2013_numbers[summer])
max_prcp_2014_summer = np.max(prcp_2014_numbers[summer])
# max_prcp_2015_summer = np.max(prcp_2015_numbers[summer]) # Year 2015 is not a complete data
print('The Maximum precipitation on summer days is (2012): ', max_prcp_2012_summer*(100/250))
print('The Maximum precipitation on summer days is (2013): ', max_prcp_2013_summer*(100/250))
print('The Maximum precipitation on summer days is (2014): ', max_prcp_2014_summer*(100/250))

# d. Median precip on non-summer rainy days
summer_rainy_leap_year = (np.arange(366)-171<90) & (np.arange(366)-171>=0)
non_summer_rainy_leap_year = np.equal(summer_rainy_leap_year, False)

summer_rainy = (np.arange(365)-171>=90) & (np.arange(365)-171<0)
non_summer_rainy = np.equal(summer_rainy, False)

median_prcp_2012_non_summer = np.median(prcp_2012_numbers[non_summer_rainy_leap_year])
median_prcp_2013_non_summer = np.median(prcp_2013_numbers[non_summer_rainy])
median_prcp_2014_non_summer = np.median(prcp_2014_numbers[non_summer_rainy])
# median_prcp_2015_non_summer = np.median(prcp_2015_numbers[non_summer_rainy]) # Dataset not complete

print('The Median on non-summer rainy days (2012) is : ', median_prcp_2012_non_summer*(100/250))
print('The Median on non-summer rainy days (2013) is : ', median_prcp_2013_non_summer*(100/250))
print('The Median on non-summer rainy days (2014) is : ', median_prcp_2014_non_summer*(100/250))
```
